```python
import pandas as pd
from sklearn.datasets import load_iris
```

```python
data = load_iris()
df = pd.DataFrame(data.data, columns=data.feature_names)

# Add target
df["target"] = data.target
```

```python
# SCHEMA VALIDATION

expected_columns = ["sepal length (cm)", "sepal width (cm)",
                     "petal length (cm)", "petal width (cm)", "target"]

schema_check = all(col in df.columns for col in expected_columns)
print("Schema Validation Passed:", schema_check)
```

```python
# DATA VALIDATION RULES

df["null_check"] = df.isnull().sum(axis=1)

df["range_validation"] = (
    (df["sepal length (cm)"] > 0) &
    (df["petal length (cm)"] > 0)
).astype(int)
```

```python
# RELEASE READINESS SCORE

df["readiness_score"] = (
    df["range_validation"]
)
```

```python
# Deployment decision
df["deployment_status"] = df["readiness_score"].apply(
    lambda x: "READY" if x == 1 else "BLOCKED"
)

print(df[["readiness_score", "deployment_status"]].head())
```
